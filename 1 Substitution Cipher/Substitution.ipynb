# Frequency Count

```python
from collections import Counter 
```

```python
#Frequency Count
res = Counter(instr[idx : idx + 1] for idx in range(len(instr) - 1)) 
for t in (sorted( res.items(), key = lambda kv:(kv[1], kv[0]) , reverse = True) ):
    print(t[0],' ',t[1])
```

```python
#Bigram Count
res = Counter(instr[idx : idx + 2] for idx in range(len(instr) - 1)) 
for t in (sorted( res.items(), key = lambda kv:(kv[1], kv[0]) , reverse = True) ):
    print(t[0],' ',t[1])
```

```python
#Trigram Count
res = Counter(instr[idx : idx + 3] for idx in range(len(instr) - 1)) 
for t in (sorted( res.items(), key = lambda kv:(kv[1], kv[0]) , reverse = True) ):
    print(t[0],' ',t[1])
```

```python
#Four Word Count
res = Counter(instr[idx : idx + 4] for idx in range(len(instr) - 1)) 
for t in (sorted( res.items(), key = lambda kv:(kv[1], kv[0]) , reverse = True) ):
    print(t[0],' ',t[1])
```

```python
# Generate Decrytped Text
def decrypter(text, dcipher):
    newtext = ''
    for c in text:
        if c in dcipher.keys():
            newtext = newtext + dcipher[c]
        else:
            newtext = newtext + c
    return newtext
```

```python
# Generate Substitution Cipher
orig_char = 'abcdefghijklmnopqrstuvwxyz'
subs_char = 'sv_qhponcml_tuba_grfdwiye_'
rep_dict = {}
for i in range(26):
    rep_dict[orig_char[i]] = subs_char[i]
    
print(instr)
print("\n")
print( decrypter(instr,rep_dict) )
```
